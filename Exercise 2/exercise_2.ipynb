```python
# Quitando datos vacios
```

```python
data_na = data.dropna()
```

```python
#Validamos
data_na.info()
```

```python
#convertir la caracteristica categoríca a numerica
#Proximidad al oceano
data_na['ocean_proximity']
```

```python
data_na['ocean_proximity'].value_counts()
```

```python
# Podriamos transformar los valores a equivalentes numericos
# 1,2,3,4,5 
# Esto puede representar un problema por que al entrenar el modelo 
# Este puede interpretar los valores mayores como un posible peso y esto afectaria nuestro modelo
```

```python
# La solución es Dummies / On-Hot Encoding
# NEAR BAY   INLAND  NEAR OCEAN
#  1            0         0
#  0            0         1
# dtype para indicarle que sea numerico
dummies = pd.get_dummies(data_na['ocean_proximity'], dtype=int)
data_na = data_na.join(dummies)
```

```python
data_na.head()
```

```python
# Luego vamos a quitar la columna ocean_proximity 
# axis = 1 para indicar que es columna
data_na = data_na.drop(['ocean_proximity'], axis=1)
```

```python
data_na.head()
```

```python
# Analisis y nuevas características
# Correlaciones 
data_na.corr()
```

```python
# sb.set para configurar parametros como el tamanio, annot para color label del valor y cmap para colores del heatmap
sb.set(rc={'figure.figsize': (15,8)})
sb.heatmap(data_na.corr(), annot = True, cmap='YlGnBu')
```

```python
# Los numeros en las etiquetas nos informan las relacional enre los datos
# LA diagonal es el mismo dato, los numeros desde el -1 a 1, 1 significa relacional perfecta y  -1 relacional inversa
# un numero que se acerca a cero o 0 significa que no hay relación
```
